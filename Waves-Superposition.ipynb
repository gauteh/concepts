```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 1. / 52
t = np.arange(0, 60, dt)
```

```python
print(t, t.shape)
```

```python
## Signal
f0 = 2 # Hz
s0 = 1.2 * np.sin(f0 * np.pi * t)
print(s0.shape)
```

```python
plt.figure()
plt.plot(t, s0)
plt.xlim([0, 4])
```

```python
f1 = 5 # Hz
a1 = 0.3 # m
phase = np.pi/3 # radians
s1 = a1 * np.sin(f1 * np.pi * t + phase)

s = s0 + s1
plt.figure()
plt.plot(t, s)
plt.xlim([0, 4])
```
